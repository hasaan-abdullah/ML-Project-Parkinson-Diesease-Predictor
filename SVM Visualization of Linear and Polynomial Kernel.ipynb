```python
#Linear Kernel
import numpy as np ##importing numpy library for numerical computations.
import matplotlib.pyplot as plt ##importing pyplot module from matplotlib for data visualization.
from sklearn import svm ##importing svm module from scikit-learn for Support Vector Machine algorithm.
import pandas as pd #for data manipulation and analysis, used for importing the data file

# load the data
data = pd.read_csv('parkinsons.csv')

#extract features (PPE and MDVP:Flo(Hz)) and labels (status) from the DataFrame as NumPy arrays
X = data[['PPE', 'MDVP:Flo(Hz)']].values 
y = data['status'].values

# create an SVM model with a linear kernel and C=100
model = svm.SVC(kernel='linear',C=100)

#fit the SVM model to the data (i.e., train the model using the input data)
model.fit(X, y)

# get the separating hyperplane
w = model.coef_[0] #get the coefficients of the separating hyperplane
a = -w[0] / w[1] #calculate the slope of the separating hyperplane
xx = np.linspace(0, 0.5) #create an array of evenly spaced values for x-coordinate
yy = a * xx - (model.intercept_[0]) / w[1] #calculate the corresponding y-coordinate values for the separating hyperplane

# loop through unique values of 'status' column
for label in np.unique(y):
    # plot data points with color based on label
    plt.scatter(X[y==label][:, 0], X[y==label][:, 1], label=label)

# plot the separating hyperplane
plt.plot(xx, yy, 'k-', label='SVM line')

#set the title, x-label, y-label and legend of the plot
plt.title('2D Scatter plot of PPE vs MDVP:Flo(Hz)')
plt.xlabel('PPE')
plt.ylabel('MDVP:Flo(Hz)')
plt.legend() #The legend() function is used to display the legend on the plot, which shows the labels associated with each data point or line on the plot.
plt.show()
```

```python
# load the data
data = pd.read_csv('parkinsons.csv')

# extract features and labels
X = data[['PPE', 'MDVP:Flo(Hz)']].values
y = data['status'].values

# create an SVM model with a polynomial kernel
model = svm.SVC(kernel='poly', degree=3,C=100)

# fit the model to the data
model.fit(X, y)

# plot the decision boundary
h = 0.02  # step size in the mesh
x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1 #calculates the minimum and maximum values of the first column (PPE) of feature matrix X with an additional margin of 0.1 on both sides.
y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1 #calculates the minimum and maximum values of the second column (MDVP:Flo(Hz)) of feature matrix X with an additional margin of 0.1 on both sides
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h)) #creates a meshgrid with xx and yy arrays to cover the feature space of PPE and MDVP:Flo(Hz) with a given step size h.
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]) #predicts the labels of the points in the feature space using the SVM model and the flattened meshgrid.
Z = Z.reshape(xx.shape) #reshapes the predicted labels to the shape of the meshgrid.
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8) #creates a filled contour plot with xx and yy coordinates, using predicted labels Z, colormap Paired and transparency alpha=0.8.

# terates through unique labels in target variable y, plots data points with color based on the label.
for label in np.unique(y): 
    plt.scatter(X[y==label][:, 0], X[y==label][:, 1], label=label)

plt.xlabel('PPE') #sets the label for the x-axis.
plt.ylabel('MDVP:Flo(Hz)') #sets the label for the y-axis.
plt.title('2D Scatter plot of PPE vs MDVP:Flo(Hz) with SVM decision boundary') #sets the title of the plot.
plt.legend() #The legend() function is used to display the legend on the plot, which shows the labels associated with each data point or line on the plot.
plt.show() #Display the Plot
```
